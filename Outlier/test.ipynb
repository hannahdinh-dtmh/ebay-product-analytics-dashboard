```python
import numpy as np
import pandas as pd
from scipy.stats import pearsonr
import seaborn as sns
import matplotlib.pyplot as plt

# Create a realistic dataset
np.random.seed(42)
n = 100

data = {
    'study_hours': np.random.normal(10, 3, n),
    'sleep_hours': np.random.normal(7, 1.5, n),
    'exercise_minutes': np.random.normal(30, 15, n),
    'stress_level': np.random.normal(5, 2, n),
    'exam_score': np.random.normal(75, 10, n)
}

# Add some correlations
data['exam_score'] = data['study_hours'] * 2 + data['sleep_hours'] * 1.5 - data['stress_level'] * 0.8 + np.random.normal(50, 5, n)
data['stress_level'] = 10 - data['sleep_hours'] * 0.5 + np.random.normal(0, 1, n)

df = pd.DataFrame(data)

print("Dataset Preview:")
print(df.head())
print(f"\nDataset shape: {df.shape}")
```
